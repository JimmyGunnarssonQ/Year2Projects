```
from numpy import *
import pandas as pd 
import matplotlib.pyplot as plt

data = pd.read_csv("data.txt", delimiter= '\t', index_col=False)
L1 = []
L2 = []
t = []

for v in data["Temperature Th, Ch A (°C) Run #1"]:
  S = list(v)
  L1.append(10*float(S[0]) + float(S[1]) + float(S[3])/10)

for k in data["Temperature Tc, Ch B (°C) Run #1"]:
  Q = list(k)
  if len(Q) == 4:
    L2.append(10*float(Q[0]) + float(Q[1]) + float(Q[3])/10)
  else:
    L2.append(float(Q[0]) + float(Q[2])/10)



for i in range(len(L1)):
  t.append(10 + 10*i)
plt.plot(t, L1, "r", label = "hot reservoir")
plt.plot(t, L2, "b", label = "cold reservoir")
plt.yticks([i for i in range(0,55,5)])
plt.xlabel("time (s)")
plt.ylabel("temperature (	°C )")
plt.title("The temperature  T (°C ) for each reservoir after t seconds")
plt.grid()
plt.legend()
plt.show()
```
